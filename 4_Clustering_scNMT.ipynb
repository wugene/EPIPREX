## read scNMT

```python
with gz.open('OUTPUT/gene_enc.pkl.gz', 'rb') as f:
    scNMT = pkl.load(f)
```

```python
with gz.open('TR_DATA/id_cells.pkl.gz', 'rb') as f:
    cells = pkl.load(f)
```

```python
cell_types = [ x.split('_')[0] for x in cells ]
cell_dict = { x:k*5 for k, x in enumerate(set(cell_types)) }
```

```python
cell_dict
```

```python
cell_i = [cell_dict[x] for x in cell_types]
```

```python
for method in ['tSNE','UMAP','PCA']:
    plot_embedding(scNMT, cell_i, method=method, save='%s.pdf' % (method), cmap='Set2', markersize=20)
```

```python
for method in ['tSNE','UMAP','PCA']:
    plot_embedding(scNMT_N, cell_i, method=method, save='%s.pdf' % (method), cmap='Set2', markersize=20)
```
